## Data Cleaning and Preprocessing

```python
df = df_orig.copy()
```

### Handle numeric columns

```python
df["discounted_price"].unique()[:10]
```

```python
df["discounted_price"] = df["discounted_price"].astype(str).str.replace('‚Çπ', '').str.replace(',', '')
df["actual_price"] = df["actual_price"].astype(str).str.replace('‚Çπ', '').str.replace(',', '')
```

```python
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors='coerce')
df["actual_price"] = pd.to_numeric(df["actual_price"], errors='coerce')
```

```python
df["discount_percentage"].unique()[:10]
```

```python
df["discount_percentage"] = df["discount_percentage"].astype(str).str.replace('%', '')
df["discount_percentage"] = pd.to_numeric(df["discount_percentage"], errors='coerce')
```

```python
df["rating"].unique()[:10]
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors='coerce')
```

```python
df["rating_count"].unique()[:10]
```

```python
df["rating_count"] = df["rating_count"].astype(str).str.replace(',', '')
df["rating_count"] = pd.to_numeric(df["rating_count"], errors='coerce')
```

### Handle missing values

```python
df.isna().sum()
```

```python
df[df["rating_count"].isna()]
```

```python
df[df["rating"].isna()]
```

```python
df_orig.at[1279, "rating"]
```

```python
df["rating_count"] = df["rating_count"].fillna(df["rating_count"].median())
df["rating"] = df["rating"].fillna(df["rating"].median())
```

## Visual EDA
- Univariate plots (histograms, bar plots)
- Bivariate plots (scatter, boxplots)
- Time series analysis (if date column exists)

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_cols = ["discount_percentage", "discounted_price", "actual_price", "rating", "rating_count"]
for col in num_cols:
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
sns.countplot(y=df["rating"], order=df["rating"].value_counts().index)
plt.title("Count of Products by Rating")
plt.show()
```

```python
# Correlation map
plt.figure(figsize=(8, 6))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f")
plt.show()
```
